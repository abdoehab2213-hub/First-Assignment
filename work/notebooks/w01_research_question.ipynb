Key Data Insights (Starter Playground):

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

# Section 3 Code: Extract 2-3 real numbers from starter dataset
if 'df' not in locals():
    df = pd.read_csv('data/raw/content_refresh_anonymized.csv') if os.path.exists('data/raw/content_refresh_anonymized.csv') else df_starter

# Number 1: Total High-Demand Pages
high_demand = df[df['impressions_90d'] >= 500]
pct_high_demand = (len(high_demand) / len(df)) * 100

# Number 2: High-Demand Pages currently declining
declining_high_demand = high_demand[high_demand['trend_direction'] == 'down']
pct_declining = (len(declining_high_demand) / len(high_demand)) * 100

# Number 3: Stale High-Demand Pages (Last updated > 180 days)
stale_high_demand = high_demand[high_demand['days_since_last_update'] >= 180]
pct_stale = (len(stale_high_demand) / len(high_demand)) * 100

print(f"1. High-Demand Pages (>=500 imp): {len(high_demand):,} ({pct_high_demand:.1f}% of total)")
print(f"2. High-Demand Declining Pages: {len(declining_high_demand):,} ({pct_declining:.1f}% of high-demand)")
print(f"3. Stale High-Demand Pages (>=180 days old): {len(stale_high_demand):,} ({pct_stale:.1f}% of high-demand)")
```
